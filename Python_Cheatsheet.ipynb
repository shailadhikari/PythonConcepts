# 2. Special Functions:
## 2.1. Print Function:

```python
# Basic print function
print("Hello World!")

# Double quotes within a string
print('"Hello" World!')
print("\"Hello\" World!")

# Single quotes within a string
print('\'Hello\' World!')
print("'Hello' World!")

# Call print function to add space between lines.
print("Hello World!")
print() # Additional space between lines
print("Howdy?")

# Write next sentence in a new line (special characters)
print("Hello World!\nThis\\ is a\t Python Cheatsheet")

# Prints with a space instead of a newline (notebook hint = shift + tab)
print("Hello", end=" ") # shift + tab
print("World")

# Print multiple variables
print("The random numbers are", 30, "and", 53)

# Control the separator between multiple items
print("Hello", "World", "from", "Python", sep=", ")

# f-strings
my_name = "Mave"
your_name = "Rick"

print(f"My name is {my_name} & your name is {your_name}")
print("My name is {} & your name is {}".format(my_name, your_name))
```

## 2.2 Input Function:

```python
# Basic input function
print(input())

# Input function with a message
print(input("Enter your name:"))

# Handling multiple inputs
x, y = input("Enter two numbers separated by a space: ").split()
print(x,"and", y)

# Reading integers from the user
user_input = input("Enter your number: ")

print()
print(user_input)
print(type(user_input))

num = int(user_input)

print()
print(num)
print(type(num))
```

# 3. Variables & Data Types

```python
#Variable declaration convention
# Valid : age, age1, _name, is_married
# Invalid: 1age, $name, name&

# Integer Data Type
age1 = 20
print(age1)
print(type(age1))

# Float Data Type
gpa = 3.7
print(gpa)
print(type(gpa))

# Boolean Data Type
is_married = True
print(is_married)
print(type(is_married))

# String Data Type
_name = "Matt"
print(_name)
print(type(_name))

# Overwriting (re-assigning) values to variables
age = 30
print(type(age))
age = "30"
print(type(age))

```

```python
x = 10
y = 3
```

```python
# Arithmetic Operators [+, -, *, /, //, %, **]
print("x + y = ", x+y) # ADDITION
print("x - y = ", x-y) # SUBSTRACTION
print("x multiplied by y: ", x*y) # MULTIPLICATION
print("x divided by y: ", x/y) # DIVISION
print("x divided to floor by y: ", x//y) # FLOOR DIVISION
print("Remainder when x is divided by y: ", x%y) #MODULO
print("x raised to the power of y: ", x**y) #POWER
```

```python
# Comparison Operators [>, <, ==, !=, >=, <=, is, is not]
print("x is less than y: ", x<y)
print("x is greater than y: ", x>y)
print("x is equal to y: ", x==y)
print("x is not equal to y: ", x!=y)
print("x is y : ", x is y)
print("x is not y : ", x is not y)
```
