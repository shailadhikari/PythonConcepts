# 2. Special Functions:
## 2.1. Print Function:
* ### Displays test, numbers, or results of expressions.
* ### Automatically adds a newline character.
* ### Can be modified by special characters i.e. \n, \t, \\, \', \", etc.

```python
# Basic print function
print("Hello World!")

# Double quotes within a string
print('"Hello" World!')
print("\"Hello\" World!")

# Single quotes within a string
print('\'Hello\' World!')
print("'Hello' World!")

# Call print function to add space between lines.
print("Hello World!")
print() # Additional space between lines
print("Howdy?")

# Write next sentence in a new line (special characters)
print("Hello World!\nThis\\ is a\t Python Cheatsheet")

# Prints with a space instead of a newline (notebook hint = shift + tab)
print("Hello", end=" ") # shift + tab
print("World")

# Print multiple variables
print("The random numbers are", 30, "and", 53)

# Control the separator between multiple items
print("Hello", "World", "from", "Python", sep=", ")

# f-strings
my_name = "Mave"
your_name = "Rick"

print(f"My name is {my_name} & your name is {your_name}")
print("My name is {} & your name is {}".format(my_name, your_name))
```

# 4. Operators
## 4.1. Arithmetic Operators

```python
a = 5
b = 2

print(f"Addition: {a} + {b} = {a+b}")
print(f"Subtraction: {a} - {b} = {a-b}")
print(f"Multiplication: {a} * {b} = {a*b}")
print(f"Power: {a} ** {b} = {a**b}")
```

In the division expression (5 = 2 x 2 + 1), the components are as follows:

* Dividend: The number to be divided, which is 5.
* Divisor: The number by which the dividend is divided, which is 2.
* Quotient: The result of the division (without considering the remainder), which is 2.
* Remainder: The part left after dividing, which is 1.

Thus, when (5) is divided by (2), the quotient is (2) and the remainder is (1).

```python
print(f"Division: {a} / {b} = {a/b}")
print(f"Floor division(Quotient): {a} // {b} = {a//b}")
print(f"Remainder: {a} % {b} = {a%b}")
```

### List of all the comparison operators in python:

### Note:
#### Difference between (a == b) and (a is b):

In Python, the id() function returns the identity of an object, which is a unique identifier for that object during its lifetime in memory. This identity is typically represented as the memory address of the object.

```python
# Difference between 'a==b' vs 'a is b' using list
a = [1, 2, 3]
b = [1, 2, 3]
print(id(a))
print(id(b))

print(a == b)
print(a is b)
```

**Conclusion:** Even though a and b contain the same values, they are two separate objects in memory. Therefore, their id() values will be different. This demonstrates that in Python, lists (and other mutable types) are objects that are stored in different memory locations even if their contents are identical.
