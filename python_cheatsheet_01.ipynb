# **4. Operators**
## **4.1. Arithmetic Operators**

````python
a = 5
b = 2

print(f"Addition: {a} + {b} = {a+b}")
print(f"Subtraction: {a} - {b} = {a-b}")
print(f"Multiplication: {a} * {b} = {a*b}")
print(f"Power: {a} ** {b} = {a**b}")
````

In the division expression (5 = 2 x 2 + 1), the components are as follows:

* Dividend: The number to be divided, which is 5.
* Divisor: The number by which the dividend is divided, which is 2.
* Quotient: The result of the division (without considering the remainder), which is 2.
* Remainder: The part left after dividing, which is 1.

Thus, when (5) is divided by (2), the quotient is (2) and the remainder is (1).

````python
print(f"Division: {a} / {b} = {a/b}")
print(f"Floor division(Quotient): {a} // {b} = {a//b}")
print(f"Remainder: {a} % {b} = {a%b}")
````

## **4.2. Comparison Operators**
### List of all the comparison operators in python:
* '**==**' : equals
* '**!=**' : not equals
* '**>**' : greater than
* '**<**' : smaller than (less than)
* '**>=**' : greater than or equal to
* '**<=**' : less than or equal to
* '**is**' : identity comparison
* '**is not**' : negative identity comparison
* '**in**' : membership test
* '**not in**' : negative membership test

````python
x = 15
y = 20
print("x is less than y: ", x<y)
print("x is greater than y: ", x>y)
print("x is equal to y: ", x==y)
print("x is not equal to y: ", x!=y)
print("x is y : ", x is y)
print("x is not y : ", x is not y)
````

### Note:
#### Difference between (a == b) and (a is b):

In Python, the id() function returns the identity of an object, which is a unique identifier for that object during its lifetime in memory. This identity is typically represented as the memory address of the object.

````python
# Difference between 'a==b' vs 'a is b' using list
a = [1, 2, 3]
b = [1, 2, 3]
print(id(a))
print(id(b))

print(a == b)
print(a is b)
````

**Conclusion:** Even though a and b contain the same values, they are two separate objects in memory. Therefore, their id() values will be different. This demonstrates that in Python, lists (and other mutable types) are objects that are stored in different memory locations even if their contents are identical.

## **5. Conditional Statements:**
---
### **5.1. If-Else Statements:**


---

* Conditional statements allow branching in code based on conditions.
* Use **if**, **elif**, and **else** keywords.
* Python uses indendation to identify code blocks (including the conditional code blocks). Best practice is to maintain consistent indentation.
* e.g.:
  ```
  if condition1:
    # code block 1 when condition1 is True
  elif condition2:
    # code block 2 when condition2 is True
  else:
    code block when condition1 and condition2 are False
  ```
* **Complex conditions:** Logical operators (and, or, not) to create complex conditions.
* **Nesting:** If-else can be nested for more complex decision-making.
* **Best Practices**:
  * Keep conditions simple and readable.
  * Use parantheses for clarity when combining conditions.








````python
# Example 5.1.1: Simple if-else statement
age = 25
threshold = 18

if(age >= threshold):
  print("You are eligible to vote!")
else:
  print("You are not eligible to vote.")
````

````python
# Example 5.1.2: Arithmetic Operators
a = 11

if(a%2 == 0):
  print(f"{a} is Even Number.")
else:
  print(f"{a} is Odd Number.")
````

````python
# Example 5.1.3.: Multiple Conditions
age = 25
has_license = True

if(age >= 18 and has_license):
  print("You are eligible to drive a car.")
elif(age >= 18 and not has_license):
  print("You need a license to drive a car.")
else:
  print("You are not eligible to drive a car.")
````

## **6. Python Loops:**
---

* Two Types of Loops available in Python:
  1.   **For** Loops
  2.   **While** Loops

*   Loop Control Statements:
    1. **break**: Exit the loop immediately.
    2. **continue**: Skip the rest of the loop and start the next iteration.

  * Best Practices for Loops Usage:
    * Keep loop bodies simple and readable
    * Avoid excessive nesting for efficiency

  * No do-while loops in Python.

### **6.1. For Loops:**

* Iterate over a sequence such as list, tuple, string and other iterables.

````python
# Example 6.1.1
for i in range(5):
  print(i, end= " ")
````

````python
# Example 6.1.2
for i in range(5, 18):
  print(i, end= ", ")
````

````python
# Example 6.1.3
for i in range(1, 12, 3):
  print(i, end= " ")
````

````python
# Example 6.1.4: Addition of a range of numbers
total = 0
for i in range(1, 11):
  total += i
print(total)

print(sum(list(range(1, 11))))
````

````python
# Example 6.1.4: Addition of a range of numbers
digit = ""
for i in range(1, 11):
  digit += str(i)
print(digit)
````


---
### **6.2. While Loops:**
---

* Execute code block while a condition is true.


````python
# Example 6.2.1: basic while loop
count = 1

while count <= 5:
  print("Num-"+str(count), end=" ")
  count += 1

print("\nCount: ", count)
````

````python
# Example 6.2.2: loop control - break

for i in range(5):
  if i==3:
    break;
  print(i, end=" ")
````

````python
# Example 6.2.2: loop control - continue

for i in range(5):
  if i==3:
    continue;
  print(i, end=" ")
````

````python
# Example: Find the sum of digits of a given number N

N = 5132

# Solution 1: Using For Loop
result1 = 0

for i in str(N) :
  result1 += int(i)

print(result1)

# Solution 2: Using While Loop
result2 = 0
while N > 0:
  result2 += N%10
  N //= 10

print(result2)
````

---
### **6.3. Nested Loops:**
---

* Loops can be placed inside other loops to create complex looping structure.


````python
# Example: Nested Loops: Multiplication Table

for i in range(1, 5):
    for j in range(1, 11):
        print(i * j, end="\t")
    print()
````

## **7. Functions in Python:**
---
* Functions:
  * Reusable pieces of code that perform a specific task.
  * Improve code reliability, modularity, and maintainability.

* Function Call:
  * Invoke a function by its name, followed by paranthese containing arguments.



````python
# Example 7.1: Function example

def function_name(arguments):
  #function body
  return
````

* Function Declaration:
  * **def:** *keyword used to declare a fucntion*
  * **function_name:** a*ny name given to a fucntion*
  * **arguments:** *any values passed to the fucntion*
  * **return (optional):** *returns value from a fucntion*

* Return Values:
  * Function can return a value using the return statement.
  * If no return statement is used, the function return None.

* Parameter vs Arguments:
  * **Parameters:** Variables defined in the fucntion's signature.
  * **Arguments:** Values passed to the fucntion for its parameters during fucntion call.

* Default parameter values:
  * Assign a default value to a parameter to make it optional.
  * Non-default parameter come before default parameters.

* Call By:
  * **Call by Value:** When the passed object is immutable.
  * **Call by Reference:** When the passed object is mutable.

````python
# Example 7.2: Default Parameter

def greet(name, greeting="Hello"):
  print(f"{greeting}, {name}!")

greet("Alice")
````

````python
# Example 7.3:
def greet():
  print("Hello")

print(greet)

print(greet())
````

````python
# Example 7.4: Type Hinting

def greet_someone(name: str, greeting="Hello") -> None:
  print(f"{greeting}, {name}!")

greet_someone("Alice")
````

````python
# Example 7.5: Airthmetic non-default parameters
def add_numbers(a: int, b: int) -> int:
  return a + b

print(add_numbers(a=2, b=3))
````

````python
# Example 7.6: Airthmetic default parameters
def add_numbers(a: int = 7, b: int = 8) -> int:
  return a + b

add_numbers()
````
