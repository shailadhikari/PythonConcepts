# Input Function
* Input Function is an in-built function in python used to take the input from user in runtime.
* Input Function takes string as an input

```python
## Code to add two numbers by using input function
num1 = input("Enter a number: ")
num2 = input("Enter another number: ")
result = int(num1) + int(num2) ## Type Casting

print(f"The sum of {num1} and {num2} is: {result}")


num3 = float(input("Enter the first number: "))
num4 = float(input("Enter the second number: "))

result = num3 + num4
print(f"The sum of {num3}, {num4} is: {result}")
```

# Type Casting
* Can change the type of a variable to another valid type

```python
num1 = "25"
print(type(num1))

num1 = int(num1)
print(type(num1))
```

```python
## Substraction
num1 = input("Enter a number: ")
num2 = input("Enter another number: ")
result = int(num1) - int(num2) ## Type Casting

print(f"The substraction of {num1} and {num2} is: {result}")
```

```python
## Multiplication
num1 = input("Enter a number: ")
num2 = input("Enter another number: ")
result = int(num1) * int(num2) ## Type Casting

print(f"The multiplication of {num1} and {num2} is: {result}")
```

```python
## Division

# Float Division
num1 = int(input("Enter a number: "))
num2 = int(input("Enter another number: "))

result = num1/num2
print(f"The float division of {num1} and {num2} is: {result}")

# Float Division
num1 = int(input("Enter a number: "))
num2 = int(input("Enter another number: "))

result = num1//num2
print(f"The floor division of {num1} and {num2} is: {result}")

# Modular Division
num1 = int(input("Enter a number: "))
num2 = int(input("Enter another number: "))

result = num1%num2
print(f"The remainder of {num1} and {num2} is: {result}")
```

```python
## Power

# Square of a number
num1 = 40
num1_square = pow(40, 2)
print(f"The square of {num1} is: {num1_square}")

# Cube of a number
num1 = 40
num1_cube = pow(40, 3)
print(f"The square of {num1} is: {num1_cube}")
```

```python
# Result of (a+b)^2
# (a+b)^2 = a^2+b^2+2*a*b

num1 = 3
num2 = 4

result = pow(num1, 2) + pow(num2, 2) + 2*num1*num2
result2 = pow((num1+num2), 2)
print(f"[Using formula]The result of (a+b)^2 for {num1} and {num2} is: {result}")
print(f"The result of (a+b)^2 for {num1} and {num2} is: {result}")
```

```python
# Result of (a+b)^2
num1 = 4.5
num2 = 7.67

result = pow((num1+num2), 2)
rounded_result = round(result,2 )

print(f"The result of (a+b)^3 for {num1} and {num2} is: {result}")
print(f"The rounded result of (a+b)^3 for {num1} and {num2} is: {rounded_result}")
```
