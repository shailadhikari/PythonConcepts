## Problem:
You have a DataFrame df with columns 'A', 'B', and 'C', and you want to sort the DataFrame based on the values in column 'B' in ascending order. Which of the following code snippets is correct?

## Code:

```python
data = {
    "A": ["IT", "HR", "IT", "Finance", "HR", "Finance"],
    "B": ["Ellie", "Bravo", "Frank", "Candy", "Aditya", "Donnie"],
    "C": [70000, 50000, 80000, 60000, 52000, 65000]
}

df = pd.DataFrame(data)
print(df)
print("--------------------")

df_sorted_1 = df.sort_values("B")
print(df_sorted_1)
print("--------------------")

df_sorted_2 = df.sort_values(by="B")
print(df_sorted_2)
print("--------------------")

df_sorted_3 = df.sort_values(by="B", ascending=True)
print(df_sorted_3)
print("--------------------")

print(df)
print("--------------------")
df.sort_values(by="B", inplace=True) ## inplace sorting in the same data frame
print(df)
print("--------------------")

```
