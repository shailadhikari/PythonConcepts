## Problem:
You have a DataFrame df with columns 'A', 'B', and 'C', and you want to sort the DataFrame based on the values in column 'B' in ascending order. Write the code snippet?


## Code:

````python
data = {
    "A": ["IT", "HR", "IT", "Finance", "HR", "Finance"],
    "B": ["Ellie", "Bravo", "Frank", "Candy", "Aditya", "Donnie"],
    "C": [70000, 50000, 80000, 60000, 52000, 65000]
}

df = pd.DataFrame(data)
print(df)
print("--------------------")

df_sorted_1 = df.sort_values("B") # Sorting the data frame column #1
print(df_sorted_1)
print("--------------------")

df_sorted_2 = df.sort_values(by="B") # Sorting the data frame column #2
print(df_sorted_2)
print("--------------------")

df_sorted_3 = df.sort_values(by="B", ascending=True) # Sorting the data frame column #3
print(df_sorted_3)
print("--------------------")

print(df)
print("--------------------")
df.sort_values(by="B", inplace=True) # Inplace sorting in the same data frame: return None
print(df)
print("--------------------")

````

## Problem:
You have a DataFrame df with columns 'A', 'B', and 'C', and you want to replace all occurrences of the value 0 in column 'A' with the mean of non-zero values in that column. Write the code snippet?

## Code:

````python
df = pd.DataFrame({
    'A': [10, 0, 20, 0, 30],
    'B': [1, 2, 3, 4, 5],
    'C': ['X', 'Y', 'Z', 'P', 'Q']
})

print(df)
print("--------------------")

result_df = df['A'].replace(0, df['A'][df['A'] != 0].mean()) # Replaces all zero values in column 'A' with the mean of the non-zero values from that same column.
print(result_df)
print("--------------------")
````

## Problem:
How can you merge two DataFrames on a common column?

## Code:

````python
df1 = pd.DataFrame({
    'common_column': [1, 2, 3],
    'A': ['a1', 'a2', 'a3']
})

df2 = pd.DataFrame({
    'common_column': [1, 2, 4],
    'B': ['b1', 'b2', 'b4']
})

merged_df_1 = pd.merge(df1, df2, on='common_column') # Merging dataframe df2 to df1
print(merged_df_1)
print("--------------------")

df3 = pd.DataFrame({
    'common_column': [1, 2, 4],
    'B': ['B1', 'B2', 'B4']
})

merged_df_2 = df1.merge(df3, on='common_column') # Merging dataframe df3 to df1
print(merged_df_2)
````

## Problem:
Which function should be used to drop a row by index from a DataFrame?

## Code:

````python
df = pd.DataFrame({
    'A': [10, 20, 30, 40, 50, 60, 70],
    'B': [100, 200, 300, 400, 500, 600, 700],
    'C': [1000, 2000, 3000, 4000, 5000, 6000, 7000]
    })

dropped_df = df.drop(0) # dropping single row

print(dropped_df)
print("--------------------")

dropped_df = df.drop(index = 1) # dropping single row by explicitly calling index
print(dropped_df)
print("--------------------")

multi_row_dropped_df = df.drop([0, 2]) # dropping multiple rows
print(multi_row_dropped_df)
print("--------------------")

df.drop([0, 2, 5], inplace=True) # dropping multiple rows in the same data frame
print(df)
````

## Problem:
Which of the following code snippets can be used to reset the index of a DataFrame?

## Code:

````python
df = pd.DataFrame({
    'A': [10, 20, 30]
}, index = [5, 6, 7])
print(df)
print("--------------------")

resetted_df = df.reset_index() # resets the index of the data frame
print(resetted_df)
print("--------------------")

resetted_df_dropped_row = df.reset_index(drop=True) # resets the index of the data frame and drops the old index
print(resetted_df_dropped_row)
print("--------------------")

df.reset_index(drop=True, inplace=True) # resets the index of the data frame in the same data frame and drops the old index
print(df)
print("--------------------")

df.reset_index(inplace=True) # resets the index of the data frame in the same data frame
print(df)
print("--------------------")

````

## Problem:
Which method is used to group a DataFrame based on the values of one or more columns?

## Code:

````python
df = pd.DataFrame({
    'Department': ['IT', 'HR', 'IT', 'Finance', 'HR'],
    'Salary': [70000, 50000, 80000, 60000, 52000]
})

grouped_df_1 = df.groupby('Department').mean() # Groups the DataFrame based on the values in the 'Department' column and get mean
print(grouped_df_1)

grouped_df_2 = df.groupby('Department').sum() # Groups the DataFrame based on the values in the 'Department' column and get sum
print(grouped_df_2)

grouped_df_3 = df.groupby('Department')['Salary'].mean() # Groups the DataFrame based on the values in the 'Department' column and get mean of the 'Salary' column
print(grouped_df_3)

````

## **Problem:**
Write a Python program using Pandas to create a DataFrame from the given data and find the total revenue and average price of items sold in each store.

```
data = {
    'Store': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Item': ['Apple', 'Banana', 'Orange', 'Grape', 'Apple', 'Banana', 'Orange', 'Grape'],
    'Price': [50, 20, 30, 60, 55, 22, 33, 65],
    'Quantity': [10, 12, 15, 16, 20, 25, 30, 35]
}
```

## **Code:**

````python
data = {
    'Store': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Item': ['Apple', 'Banana', 'Orange', 'Grape', 'Apple', 'Banana', 'Orange', 'Grape'],
    'Price': [50, 20, 30, 60, 55, 22, 33, 65],
    'Quantity': [10, 12, 15, 16, 20, 25, 30, 35]
}

df = pd.DataFrame(data)

# Create Total Revenue column
df['Total Revenue'] = df['Price'] * df['Quantity']
print(df)
print("--------------------")
df_s = df.groupby('Store')
print(df_s)
print("--------------------")

df_grouped = df.groupby('Store').agg(
    Total_Revenue=('Total Revenue', 'sum'),
    Average_Price=('Price', 'mean')
)

print(df_grouped)
print("--------------------")
````

## **Problem:**
Write a Python program using Pandas to create a DataFrame from the given data and calculate the total amount spent by each customer.

```
data = {
    'Customer': ['Alice', 'Bob', 'Alice', 'Alice', 'Bob', 'Bob', 'Alice', 'Bob'],
    'Item': ['Pen', 'Pencil', 'Notebook', 'Eraser', 'Pen', 'Pencil', 'Notebook', 'Eraser'],
    'Price': [10, 5, 50, 20, 10, 5, 50, 20],
    'Quantity': [3, 4, 2, 5, 10, 6, 1, 2]
}
```

## **Code:**

````python
data = {
    'Customer': ['Alice', 'Bob', 'Alice', 'Alice', 'Bob', 'Bob', 'Alice', 'Bob'],
    'Item': ['Pen', 'Pencil', 'Notebook', 'Eraser', 'Pen', 'Pencil', 'Notebook', 'Eraser'],
    'Price': [10, 5, 50, 20, 10, 5, 50, 20],
    'Quantity': [3, 4, 2, 5, 10, 6, 1, 2]
}

df = pd.DataFrame(data)

df['Total Amount'] = df['Price'] * df['Quantity'] # Create new column wiht the total amount spent in that row.

print(df)
print("--------------------")

df_grouped = df.groupby('Customer').agg(
    Total_Amount=('Total Amount', 'sum')
) # Group the data frame by 'Customer' and aggregate the sum of the total amount for the groups.
print(df_grouped)
````

## **Problem:**
Write a Python program using Pandas to create a DataFrame from the given data and find the month with the highest average temperature.
```
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Temperature': [20, 22, 25, 27, 30, 32, 35, 34, 30, 28, 24, 21]
}
```

## **Code:**

````python
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Temperature': [20, 22, 25, 27, 30, 32, 35, 34, 30, 28, 24, 21]
}

df = pd.DataFrame(data)

print(df)

print("--------------------")

max_temp_row= df.loc[df['Temperature'].idxmax()] # Gives the row with max temperature

print(max_temp_row)
print("--------------------")

max_month = df.loc[df['Temperature'].idxmax(), 'Month'] # Find the index of the maximum value in the 'Temperature' column
print(max_month)
````

## **Description:**
* **.loc** in Pandas is used to select rows and columns by their labels (names).
* Here, `df.loc[df['Temperature'].idxmax(), 'Month']` picks the row for the max 'Temperature' and returns corresponding value in the column 'Month'.
