# Plotting with Matplotlib

Integrates well with numpy and pandas

* Basic Plot Types:
* Line Chart: Good for Time Series Data
* Scatter Chart: Good when looking for correlations between the data points
* Bubble Chart: Similar to scatter chart with bubble size as another data insight
* Histogram: Used mostly for distribution analysis of the dataset
* Bar Chart: Compare quantities across categories
* Heat Map

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from networkx.algorithms.bipartite.basic import color
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [10000, 20000, 40000, 15000, 45000, 50000]

plt.plot(months, sales, marker = '*', color = 'b', linewidth = 5, linestyle = '--')
plt.show()
```
