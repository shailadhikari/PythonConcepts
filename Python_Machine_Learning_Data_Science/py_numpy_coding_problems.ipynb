## Problem:
Which function can be used to concatenate two NumPy arrays horizontally?

````python
A = np.array([[1,2,3],[4,5,6]])
B = np.array([[7,8,9],[10,11,12]])
print(np.hstack((A, B))) # horizontal concatination
print("______________")
print(np.vstack((A,B))) # vertical concatination
print("______________")

# Same can be done using concatenate and an axis
print(np.concatenate((A,B), axis=1)) # horizontal concatination
print("______________")
print(np.concatenate((A,B), axis=0)) # vertical concatination
````

## Problem:
Using NumPy, create a 5x5 identity matrix and add a scalar value of 5 to it. Multiply the resulting matrix with a 5x1 column vector filled with random integers between 1 and 10. Print the resulting vector.

````python
arr = np.eye(5) # Creates a 5X5 identity matrix
print(arr)
print("-------------------------------------")

arr += 5
print(arr)
print("-------------------------------------")

col_vec = np.random.randint(1,11, size=(5,1))
print(col_vec)
print("-------------------------------------")

print(np.dot(arr, col_vec))
````

* The **dot product**, also known as the scalar product, is a fundamental operation in linear algebra.
  * It takes two equal-length sequences of numbers (vectors) and returns a single number (a scalar).
* **np.dot()** :  used for calculating the dot product.
  * For 1D arrays, it calculates the **inner product** of vectors.
  * For 2D arrays, it is equivalent to **matrix multiplication**.
  * For higher-dimensional arrays, it's a sum product over the last axis of a and the second-to-last axis of b.

## Problem:
Write a Python program using NumPy to create a 10x10 matrix filled with random integers between 1 and 100. Find the minimum and maximum values of the matrix along with their indices.

````python
A =  np.random.randint(1, 101, size =(10,10))
print(A)

min_val = np.min(A)
max_val = np.max(A)
print(min_val, max_val)

min_val_idx = np.argmin(A) # Returns the list index of min val
max_val_idx = np.argmax(A) # Returns the list index of max val
print(min_val_idx, max_val_idx)

min_idx = np.unravel_index(min_val_idx, A.shape) # Unravels the exact matrix index for min val
max_idx = np.unravel_index(max_val_idx, A.shape) # Unravels the exact matrix index for max val
print(min_idx, max_idx)
````

Write a Python program using NumPy to create a 5x5 matrix filled with random floats between 0 and 1. Replace the maximum value in the matrix with 0 and print the original and modified matrices.

````python
A = np.random.random(size=(5,5))
print(A)

max_idx = np.unravel_index(np.argmax(A), A.shape)
print(max_idx)

A[max_idx] = 0
print(A)
````
