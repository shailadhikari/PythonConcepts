## Problem:
Which function can be used to concatenate two NumPy arrays horizontally?

````python
A = np.array([[1,2,3],[4,5,6]])
B = np.array([[7,8,9],[10,11,12]])
print(np.hstack((A, B))) # horizontal concatination
print("______________")
print(np.vstack((A,B))) # vertical concatination
print("______________")

# Same can be done using concatenate and an axis
print(np.concatenate((A,B), axis=1)) # horizontal concatination
print("______________")
print(np.concatenate((A,B), axis=0)) # vertical concatination
````

Write a Python program using NumPy to create a 10x10 matrix filled with random integers between 1 and 100. Find the minimum and maximum values of the matrix along with their indices.

````python
A =  np.random.randint(1, 101, size =(10,10))
print(A)

min_val = np.min(A)
max_val = np.max(A)
print(min_val, max_val)

min_val_idx = np.argmin(A)
max_val_idx = np.argmax(A)
print(min_val_idx, max_val_idx)

min_idx = np.unravel_index(min_val_idx, A.shape)
max_idx = np.unravel_index(max_val_idx, A.shape)
print(min_idx, max_idx)
````

Write a Python program using NumPy to create a 5x5 matrix filled with random floats between 0 and 1. Replace the maximum value in the matrix with 0 and print the original and modified matrices.

````python
A = np.random.random(size=(5,5))
print(A)

max_idx = np.unravel_index(np.argmax(A), A.shape)
print(max_idx)

A[max_idx] = 0
print(A)
````
