## **Lists:**


---



* Lists are **mutable** and **ordered** collection of elements.
* Lists allow the data storage in a **sequential** manner.
* In python, lists can be easily created using **square brackets []**.


### **Characteristics of a Python List:**

* **Ordered:** *Lists maintain the order in which the data is inserted.*
* **Mutable:** *In list element(s) are changeable. It means that we can modify the items stored within the list.*
* **Heterogeneuos:** *Lists can store elements of various data types.*
* **Dynamic:** *List can expand or shrink automaticaly to accomodate the items accordingly.*
* **Duplicate Elements:** *Lists allow us to store duplicate data.*


### **Basic Operations:**
  * **Accesing elements:** `list_name[index]`
  * **Modifying elements:** `list_name[index] = new_value`
  * **Length of the list:** `len(list_name)`
  * **Concatenation:** `list1 + list 2`
  * **Repetition:** `list_name * n` (Concatenating list to itself n number of times)


#### **Slicing:**
Extracting a subset of elements from a list based on specified indices or slice notation.

**`new_list = original_list[start:stop:step]`**

* **start:** The index of the first element to include. Default = 0
* **end:** The index of the first element to exclude. Default = len(list)
* **step:** The step size for the increment. Default = 1.






### **Iteration and List Comprehensions:**
* Iterating over a list:
  * Use a for loop to iterate and process each element in the list.

* List comprehensions:
  * Concise way to create a new list using a single line of code.
  * Syntax:
  [*expression* **for** *item* **in** *iterable* **if** *condition*]
  
* Common use cases:
  * Dynamic data storage and data manipulation due to mutable nature.
  * Implementing data structure and algos (stack, queue, sorting, seraching, etc.)

```python
# Exercise : Creating a simple list
my_list = [1, 3, "Maverick", True, 'a']

print(my_list)
print(type(my_list))
```

```python
# Exercise: Accessing the list element
print(my_list[0])
print(my_list[3])

print(my_list[len(my_list) - 1]) # last element
```

### **Top Methods of Python Lists:**
  * **append(item):** add an item to the end of the list.
  * **extend(iterable):** add items of lists and others iterables to the end of the list.
  * **insert():** inserts an item at the specified index.
  * **remove(index):** removes item present at the given index.
  * **pop(index):** returns and removes item present at the given index.
  * **clear():** removes all items from the list.
  * **index(item):** returns the index of the first matched item.
  * **count(index):** returns the count of the specified item in the list.
  * **sort():** sort the list in ascending/descending order.
  * **reverse():** reverses the item of the list.
  * **copy():** returns the shallow copy of the list.

```python
# Exercise: Appending a list by an item
fruits = ['apple', 'banana', 'cherry']
fruits.append("orange")
print(fruits)
```

```python
# Exercise: Extend a list by another iterable
#Extending a list by a list
new_fruits = ['kiwi', 'mango']
fruits.extend(new_fruits) # doesn't work for tuple as it modifies.
print(fruits)

# Extending a list by a set
new_set = {'s_avocado', 's_durian'}
fruits.extend(new_set)
print(fruits)
```

```python
# Exercise: Remove an element using remove()
fruits.remove('banana')
print(fruits)
```

```python
# Exercise: Removing and retrieving the removed element using pop()
fruits.pop(0)
print(fruits)
```

### **NOTE:** Strings are nothing but a list of characters.

```python
string_1 = "some example text"
print(f"{string_1[0]} is the first letter of {string_1}")
print("Last letter:",string_1[len(string_1)-1])
```

## 1.2.List Slicing:
### Lists can be sliced using the syntax list_name[start:stop:step].
* ### start : index of the list where slicing starts
* ### stop : index of the list before which slicing ends
* ### step : stepping to nth item within the range start to stop

```python
print(string_1[1:6]) # Characters from pos 1 tp pos 5

# string slicing from beginning
print(string_1[0:6]) # Naive way to get characters from beginning to pos 5
print(string_1[:6]) # More Pythonic way to do the above

# string slicing till the end
print(string_1[5:len(string_1)]) # Naive way to get characters from pos 5 till the end
print(string_1[5:]) # More Pythonic way to do the above

# string slicing to get entire string
print(string_1[:])

string_2 = "0123456789"
print(string_2[2:8])
print(string_2[2:8:]) # Using step with default 0 value
print(string_2[2:8:2]) # Using step value of 2

# Out of bound handling by slicing
print(string_2[2:30])
print(string_2[-30:4])
print(string_2[2:8:100])
```

```python
# Logical Operators with lists
list1=[2,5,3]
list2=[3,1]
list3=[]
list3=list1&list2
print(list3)

list3=list1|list2
print(list3)
```

```python
set_a = {1,2,3,4,5}
set_b = {4,5,6,7,8}

print(set_a.union(set_b))
print(set_b.union(set_a))
print(set_a.intersection(set_b))
print(set_b.intersection(set_a))
print(set_a.difference(set_b))
print(set_b.difference(set_a))

```

```python
# Logical Operators with sets
set1={2,5,3}
set2={3,1}
set3={}
set3=set1&set2
print(set3)

set3=set1|set2
print(set3)
```
