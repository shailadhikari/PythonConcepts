# Print Function Demo

* Prints to console
* Useful for debugging
* Print by default use new line ('\n') as the end separator
* Use 'end' parameter of print function to declare the ending character
* Use 'sep' parameter to separate multiple args of print function
> Formating in print
* F String: Use f string print function to format string based on tokens : print(f"*")
* Format(): Use format() function of string to format strings :  sequence matters; extra args of format are ignored
* Format() with index: Use format function of string with index to format strings while indexing the args
* % String: Use % string in print function to format: print('%s..'%())

```python
name = 'Shailesh'
name

age = 38
age

nationality = "Indian"
nationality

name, age, nationality
```

```python
print(name)
print('\n')
print(age)
print(nationality)
```

```python
print(name, end = '\n') # Default print behavior to end with a new line
```

```python
print(name, end = ',')
print(age, end = ',')
print(nationality, end = '\n')
```

```python
print(name, age, nationality)
```

```python
print(name, age, nationality, sep = ', ')
```

```python
print("My name is "+name+". I am "+str(age)+" years old.")
```

> F String

```python
employee = 'Harshita'
company = 'Blackrock'
salary = 100000
location = 'Delaware'

print(f'{employee} is working in {company} at {location} receiving compensation of {salary}.')
```

> Format


```python
print('{} is working in {} at {} receiving compensation of {}'.format(employee, company, location, salary))
```

> Format with index

```python
print('{a} is working in {b} at {c} receiving compensation of {d}'.format(b=company, c=location, d=salary, a=employee))
```

> % string format

```python
print('%s is working in %s at %s receiving compensation of %s'%(employee, company, location, salary))
```
