## Problem: 2 Sum In A Sorted Array

```python
def pair_sum_sorted_array(numbers, target):
    """
    Args:
     numbers(list_int32)
     target(int32)
    Returns:
     list_int32
    """

    start = 0
    end = len(numbers)-1

    while start < end :
        total = numbers[start] + numbers[end]
        if total == target:
            return [start, end]
        elif total < target:
            start+=1
        else:
            end-=1

    return [-1,-1]
```
