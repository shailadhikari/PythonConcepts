## Problem: 2 Sum In A Sorted Array

```python
def pair_sum_sorted_array(numbers, target):
    """
    Args:
     numbers(list_int32)
     target(int32)
    Returns:
     list_int32
    """

    start = 0
    end = len(numbers)-1

    while start < end :
        total = numbers[start] + numbers[end]
        if total == target:
            return [start, end]
        elif total < target:
            start+=1
        else:
            end-=1

    return [-1,-1]
```

## Problem: Majority Element

```python
def majority_element(nums):
    """
    Args:
     nums(list_int32)
    Returns:
     int32
    """
    nums.sort()
    print(nums)
    curr = nums[0]
    count = 1
    max_count = count
    result = nums[0]

    for i in range(1,len(nums)):
        print(nums[i], count, max_count, result)
        if nums[i] == curr:
            count += 1
            if max_count < count:
                result = nums[i]
        else:
            max_count = max(max_count, count)
            curr = nums[i]
            count = 1

    if max_count < count:
        result = nums[-1]

    return result

majority_element([0, 0, 1, 1, 0, 0])
```

## Problem: Check If An Array Contains Duplicates

```python

def check_if_array_contains_duplicate(nums):
    """
    Args:
     nums(list_int32)
    Returns:
     bool
    """
    nums.sort()
    print(nums)
    prev = nums[0]
    for i in range(1, len(nums)):
        print(prev, nums[i])
        if prev == nums[i]:
            print(prev, nums[i])
            return True
        prev = nums[i]
    return False

check_if_array_contains_duplicate([10, 30, 10, 20])
```

## Problem: Intersection Of Two Sorted Arrays

```python
def get_intersection(numbers1, numbers2):
    """
    Returns the intersection of two sorted integer lists,
    including duplicates, in sorted order.
    If no intersection, returns [-1] to match your current convention.
    """
    result = sorted(set(numbers1)&set(numbers2))
    return result if len(result) else [-1]


get_intersection([5, 9, 11],[1, 4, 10, 12])
```
