## Problem:
### Write a Python program that prompts the user for two numbers and divides them. Handle any exceptions that may arise from the division.

```python
entry = input("Enter two numbers separated by space: ")
numbers = entry.split()

try:
    if len(numbers) != 2:
        raise ValueError("Please enter exactly two numbers.")

    number1 = float(numbers[0])
    number2 = float(numbers[1])

    answer = number1 / number2
    print("Result:", answer)

except ZeroDivisionError:
    print("Error: Cannot divide by zero.")

except ValueError as ve:
    print("Error:", ve)

except Exception as e:
    print("Unexpected error:", e)
```
