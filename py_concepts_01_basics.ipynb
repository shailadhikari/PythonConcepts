```python
from datetime import date
from datetime import datetime

print("Welcome to the python world")
# int : 3, 5, 7, 9, 10, 50, 72
print(type(12))
# float : 12.56, 34.3, 92.5896
print(type(12.56))
# str : "1", '1', 'rocky', '5'
print(type('rocky'))
#print(type("rocky"))

#date : '2025-02-12'
d = date(2025, 2, 12)  # yyyy, mm, dd
print(d)
print(type(d))

#datetime : '2025-11-03 23:33:48.798629'
now = datetime.now()
print(now)
print(type(now))


```
