# Lists
* In-built data type in python.
* Used to store more than one element.
* List can store both homogeneous and heterogeneous elements
* List is ordered and mutable.
* List ordering and slicing operation structure: ls[ : : ] >> ls[lower_limit : upper_limit : direction] where default direction is 1

```python
## Creating a List
homo_ls = ['Apple', 'Banana', 'Eggs', 'Bread', 'Veggies']
# Printing list
print(homo_ls)
# Length of the list
print(len(homo_ls))
# Accessing element of a list
print(homo_ls[0])
# Slicing of the list
print(homo_ls[0:2]) # ls[0:2] = ls[:2] = ls[0-x:2]
print(homo_ls[:2])
print(homo_ls[-5:2])
print(homo_ls[2:]) # ls[2:] = ls[2:len(ls)] = ls[2:len(ls)+x]
print(homo_ls[2:len(homo_ls)])
print(homo_ls[2:100])
```

```python
## Creating a heterogeneous list
hetero_ls = [2, 5, 'Rocky', 'Philadelphia', 6.1, 3+4j]
print(hetero_ls)
#for i in hetero_ls:
print(f"The type of element {hetero_ls[len(hetero_ls)-1]} is {type(hetero_ls[len(hetero_ls)-1])}")
# Negative indexing in lists
print(hetero_ls[-len(hetero_ls)]) # First element : hetero_ls[-len(hetero_ls)] = hetero_ls[0]
print(hetero_ls[-1]) #Last element : hetero_ls[-1] = hetero_ls[len(hetero_ls)-1
# Slicing the list using negative index
print(hetero_ls[:-2])
# Reversing the list
print(hetero_ls[::-1]) # ls [lower_limit : upper_limit : direction]; direction = -1 means reverse order
# Slicing with direction
print(hetero_ls[-1:-4:-1])
print(hetero_ls[4:0:-1])
print(hetero_ls[0:4:-1]) # This gives empty list coz 0 to 4 is not valid indexing for reverse direction

# Appending an element to a list
hetero_ls.append('Appended element') # Appends only one element at a time
print(hetero_ls)
#Appending list to a list
hetero_ls.append(homo_ls)
print(hetero_ls)
print(hetero_ls[-1][3:0:-1])
# Appending/extending all the list elements into another list
ext_ls = [12, 'java', 'spring', 5.5, 'kafka']
ext_ls.extend(homo_ls)
print(ext_ls)

# Concatenating the list: works similarly as the extend
concat_ls = [4, 'Marshall', 'Firefighter']
concat_ls = concat_ls + homo_ls
print(concat_ls)
```

```python
# List operations 
ls = [5, 2, 7, 9, 1, 15, 21, 7]
print(ls)
print(sum(ls))
print(min(ls))
print(max(ls))
print(ls.count(7)) #Returns the count of the passed element in the list

# Updating elements of a list
ls[2] = 5
print(ls)
ls[0] = 'Skye'
print(ls)

# Inserting elements in the list
ls.insert(1, 'Rubble')
print(ls)

# Removing elements from a list
ls.remove('Skye') # removes the first occurence of element
print(ls)

ls.pop(0) # pops the element at the index
print(ls)

del(ls[0])
print(ls)

del ls[0]
print(ls)

# Sorting
ls.sort() # Heterogeneous lists can't be sorted: sort() throws an error in case of heterogeneous lists
print(ls)

# Clearing the list
ls.clear()
print(ls)

# Deleting the list
del(ls) # del ls or del(ls) shall delete the list
print(ls) # After deletion this line shall throw an error of ls being undefined
```
